# Conditional Probability

Conditional probability is the likelihood of an event occurring, given that another event has already happened. It is not the same as the probability of two events happening together; instead, it focuses on the dependency between them. This concept helps us refine our predictions by incorporating new information, making it an essential tool in statistics and everyday decision-making.

## Importing Libraries

```python
try:
    import pandas as pd
    import random
    import seaborn as sns
    import matplotlib.pyplot as plt
    print('Importing successful')
    
except ImportError:
    pass
```

## Function to create some Data on a Email Marketing Campaign

clicked_email_prob: The probability of an email being effective enough to get a click.

purchase_prob_given_click: The conversion rate for customers who showed interest by clicking.

purchase_prob_no_click: The conversion rate for all other customers who made a purchase without a direct marketing interaction.

```python
# Function to run the simulation and return results as a dictionary
def run_simulation(total_customers, clicked_email_prob, purchase_prob_given_click, purchase_prob_no_click):
    customers = []
    for i in range(1, total_customers + 1):
        clicked_email = 'Yes' if random.random() < clicked_email_prob else 'No'
        
        if clicked_email == 'Yes':
            made_purchase = 'Yes' if random.random() < purchase_prob_given_click else 'No'
        else:
            made_purchase = 'Yes' if random.random() < purchase_prob_no_click else 'No'

        customers.append({
            'customer_id': i,
            'clicked_email': clicked_email,
            'made_purchase': made_purchase
        })

    df = pd.DataFrame(customers)

    clicked_email_count = len(df[df['clicked_email'] == 'Yes'])
    made_purchase_count = len(df[df['made_purchase'] == 'Yes'])
    both_events_count = len(df[(df['clicked_email'] == 'Yes') & (df['made_purchase'] == 'Yes')])
    
    # Calculate probabilities
    p_a = made_purchase_count / total_customers
    p_b = clicked_email_count / total_customers
    p_a_and_b = both_events_count / total_customers
    
    # Handle division by zero for P(A|B)
    p_a_given_b = p_a_and_b / p_b if p_b > 0 else 0.0

    return {
        'P(A)': p_a,
        'P(B)': p_b,
        'P(A|B)': p_a_given_b
    }
```

## Set the nubmer of Customers for our scenarios

```python
total_customers = 1000  
```

## Set the Paramters for the two scenarios

```python
# Scenario 1: Baseline Campaign
results_baseline = run_simulation(
    total_customers=total_customers,
    clicked_email_prob=0.3,
    purchase_prob_given_click=0.4,
    purchase_prob_no_click=0.1
)

# Scenario 2: Optimized Campaign
results_optimized = run_simulation(
    total_customers=total_customers,
    clicked_email_prob=0.5,
    purchase_prob_given_click=0.6,
    purchase_prob_no_click=0.1
)
```

## Plot the two different results in Seaborn

```python
data = {
    'Metric': ['Overall Purchase Rate (P(A))', 'Email Click Rate (P(B))', 'Purchase Given Click (P(A|B))'] * 2,
    'Probability': [
        results_baseline['P(A)'], results_baseline['P(B)'], results_baseline['P(A|B)'],
        results_optimized['P(A)'], results_optimized['P(B)'], results_optimized['P(A|B)']
    ],
    'Campaign': ['Baseline'] * 3 + ['Optimized'] * 3
}

plot_df = pd.DataFrame(data)

# --- Create Seaborn Grouped Bar Chart ---
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

barplot = sns.barplot(
    x='Metric',
    y='Probability',
    hue='Campaign',
    data=plot_df,
    palette='viridis',
    errorbar=None  # Remove error bars for clarity
)

# Add value labels on top of the bars
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():.2f}', 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha='center', va='center', 
                   xytext=(0, 9), 
                   textcoords='offset points')

plt.title('Comparison of Probabilities: Baseline vs. Optimized Campaign')
plt.ylabel('Probability')
plt.xlabel('Probability Metric')
plt.ylim(0, 1.0)
plt.show()
```

The generated chart visually represents the two examples we discussed, providing a clear comparison of the key metrics.

The "Baseline Campaign" bars represent the results from our first simulation, showing the probabilities when the email is moderately effective.

The "Optimized Campaign" bars represent the results from the second simulation, where we increased the click-through rate and the conditional purchase probability.

The chart makes it easy to see the impact of the changes:

P(B): The "Email Click" bar for the optimized campaign is significantly taller, showing the higher click-through rate.

P(A|B): The "Purchase Given Click" bar for the optimized campaign is much higher, visually representing the stronger conversion power of the new campaign.

P(A): The "Overall Purchase" bar is also higher for the optimized campaign, demonstrating how the improvements in the campaign ultimately led to a better overall business outcome.

```python
# --- Scenario 1: Baseline Campaign ---
print("--- Baseline Campaign Calculations ---")
clicked_email_prob_1 = 0.3
purchase_prob_given_click_1 = 0.4
purchase_prob_no_click_1 = 0.1

# P(B) is directly the clicked email probability
p_b_1 = clicked_email_prob_1
# P(A|B) is directly the purchase probability given a click
p_a_given_b_1 = purchase_prob_given_click_1
# P(A|~B) is directly the purchase probability with no click
p_a_given_not_b_1 = purchase_prob_no_click_1
# P(~B) is the complement of P(B)
p_not_b_1 = 1 - p_b_1

# Calculate the Overall Purchase Rate (P(A)) using the Law of Total Probability
p_a_1 = (p_a_given_b_1 * p_b_1) + (p_a_given_not_b_1 * p_not_b_1)

print(f"P(B) (Email Click Rate): {p_b_1:.2f}")
print(f"P(A|B) (Purchase Given Click): {p_a_given_b_1:.2f}")
print(f"P(A) (Overall Purchase Rate): {p_a_1:.2f}")

print("\n" + "="*40 + "\n")

# --- Scenario 2: Optimized Campaign ---
print("--- Optimized Campaign Calculations ---")
clicked_email_prob_2 = 0.5
purchase_prob_given_click_2 = 0.6
purchase_prob_no_click_2 = 0.1

# P(B) is directly the clicked email probability
p_b_2 = clicked_email_prob_2
# P(A|B) is directly the purchase probability given a click
p_a_given_b_2 = purchase_prob_given_click_2
# P(A|~B) is directly the purchase probability with no click
p_a_given_not_b_2 = purchase_prob_no_click_2
# P(~B) is the complement of P(B)
p_not_b_2 = 1 - p_b_2

# Calculate the Overall Purchase Rate (P(A)) using the Law of Total Probability
p_a_2 = (p_a_given_b_2 * p_b_2) + (p_a_given_not_b_2 * p_not_b_2)

print(f"P(B) (Email Click Rate): {p_b_2:.2f}")
print(f"P(A|B) (Purchase Given Click): {p_a_given_b_2:.2f}")
print(f"P(A) (Overall Purchase Rate): {p_a_2:.2f}")
```
