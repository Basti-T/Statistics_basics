## Uniform Distribution 
### All values probability is in the same range

```python
np.random.seed(23)
```

```python
values_uniform = np.random.uniform(-100, 100, 100000)
```

```python
plt.hist(values_uniform, 100)
plt.show()
```

```python
mean_normal = 0
std_normal = 33
```

```python
values_normal = np.random.normal(mean_normal, std_normal, 100000)
```

```python
plt.hist(values_normal, 100)
plt.show()
```

```python
scale = 50
```

```python
values_exp = np.random.exponential(scale, 10000)
```

```python
plt.hist(values_exp, 100)
plt.show()
```

```python
n = 200
p = 0.5
```

```python
values_binom = np.random.binomial(n, p, 10000)
```

```python
plt.hist(values_binom, 100)
plt.show()
```

```python
lam = 100
```

```python
values_poisson = np.random.poisson(lam, 100000)
```

```python
plt.hist(values_poisson, 100)
plt.show()
```
